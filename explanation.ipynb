Tarinan mukaan biologit menivät metsään luokittelemaan puita. He mittasivat kaiken mahdollisen ja antoivat oman arvionsa puun lajista. Tehtäväsi (tutkimuskysymykset) on
- arvioida kuka biologeista luokitteli puut parhaiten (eli tässä tapauksessa dataan perustuen).
- tunnistaa datasta parhaiten luokittelevat attribuutit, jotta biologien ei tarvitse mitata turhia asioita excursioillaan

Datassa seuraavat attribuutit
  length, puun pituus
  diameter, puun halkaisija
  cones, käpyjen määrä
  leaves, lehtien/havujen määrä
  bark_thickness, kaarnan paksuus
  seed_size, siemenen koko
  fungi_present, oliko lähistöllä sieniä
  color_guess, arvio lehtien/havujen väristä
  soil_pH, maan pH
  collector_id, luokittelijan tunniste
  collector_guess, luokittelijan oma arvio puun luokasta

1. Agreement-based methods (quick baseline)

You can compare each biologist to the others:

Compute pairwise agreement (e.g., % of trees where two biologists gave the same label)
Or use something like Cohen’s kappa (for pairs) or Fleiss’ kappa (for many raters)

👉 Idea:
A “good” classifier tends to agree with others more often.

⚠️ Limitation:

If most people are wrong in the same way, this rewards the wrong answer.

2. Majority vote as a proxy truth
For each tree, take the most common label as a “pseudo ground truth”
Then score each biologist against that

👉 Simple and often surprisingly effective.

⚠️ Limitation:

Assumes the majority is usually correct (not always true)
