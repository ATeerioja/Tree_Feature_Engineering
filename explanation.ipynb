Tarinan mukaan biologit menivät metsään luokittelemaan puita. He mittasivat kaiken mahdollisen ja antoivat oman arvionsa puun lajista. Tehtäväsi (tutkimuskysymykset) on
- arvioida kuka biologeista luokitteli puut parhaiten (eli tässä tapauksessa dataan perustuen).
- tunnistaa datasta parhaiten luokittelevat attribuutit, jotta biologien ei tarvitse mitata turhia asioita excursioillaan

Datassa seuraavat attribuutit
  length, puun pituus
  diameter, puun halkaisija
  cones, käpyjen määrä
  leaves, lehtien/havujen määrä
  bark_thickness, kaarnan paksuus
  seed_size, siemenen koko
  fungi_present, oliko lähistöllä sieniä
  color_guess, arvio lehtien/havujen väristä
  soil_pH, maan pH
  collector_id, luokittelijan tunniste
  collector_guess, luokittelijan oma arvio puun luokasta

Haluan nähdä työssä tilastotieteellistä otetta, eli datan kuvailua & siivousta, jakaumat, datan tunnusluvut, p-arvot, jne pitäisi ilmoittaa. Työn rakenteeksi 1) datan kuvailu + preprosessointi 2) tutkimus (mahdollinen feature engineering + käyttämäsi testit/menetelmät + niiden tulokset) 3) loppupäätelmät. Kirjoita loppuraportti (palauta PDF), jossa vähintään mainitut osat; muilta osin se voi olla vapaamuotoinen.

Puilla on jokin oikea luokka ja tutkijat ovat esittäneet oman parhaan arvauksensa perustuen tehtyyn metsäreissuun. Tutkijat perustivat siis arvauksensa asiantuntijuuteen, kokemuksen ja substanssiosaamiseen ja heille tuli hieman kiistaa kuka luokitteli puut parhaiten.
